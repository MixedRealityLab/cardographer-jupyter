## Read CSV

Read the merged export with details => 1 group per column. 
Details is JSON array of cards, each a JSON array of box/frame text/titles.
(The order seems arbitrary.)

```python
import pandas as pd
mrdetailsdf = pd.read_csv('data/Initial MR 22 cards analysis Card Use - merged and details.csv', 
                          index_col=0, dtype=str, keep_default_na=False)
mrdetailsdf
```

## How many groups use a card in a box?

What are the boxes? (add more...)
- "Relevant challenges for Idea 1", "Relevant challenges for Idea 2"
- "Relevant questions for Idea 1", "Relevant questions for Idea 2"

Parse the JSON and flatten the array to it is just a list of box/frame names in each cell:

```python
import json
def flatten(t):
    return [item for sublist in t for item in sublist]
flatdf = mrdetailsdf.applymap(lambda x: flatten(json.loads('[]' if x=='' else x)))
flatdf.head()
```

### Relevant Challenges

First, is it a "Relevant challenges for Idea 1" or "...2"?

```python
rel_chal = flatdf.applymap(lambda x: 1 if 'Relevant challenges for Idea 1' in x or 'Relevant challenges for Idea 2' in x else 0)
rel_chal
```

Just the count:

```python
rel_chal.agg(sum, axis="columns")
```
